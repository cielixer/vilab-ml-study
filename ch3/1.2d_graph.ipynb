```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
# % for shell commands
%matplotlib inline # inline plots in viewer
```

## Function to plot
$$f(x, w) = (x-w) \times x \times (x+2)$$


```python
# define function to draw
def f(x, w):
    return (x-w) * x * (x+2)

# evenly space 100 numbers in interval [-3, 3]
x = np.linspace(-3, 3, 100)

plt.plot(x, f(x, 2), color='black', label='$w=2$') # draw graph f(x, 2)
plt.plot(x, f(x, 1), color='blueviolet', label='$w=1$') # draw graph f(x, 1)
plt.legend(loc="upper left") # add legned
plt.ylim(-15, 15) # y value limit
plt.title('$f(x)$') # set title(latex style)
plt.xlabel('$x$') # set x axis label (latex style)
plt.ylabel('$y$') # set y axis label (latex style)
plt.grid(True) # draw grid
plt.show() # display plot
```

```python
# plot many graphs using subplot
plt.figure(figsize=(10, 3)) # width and height of figure
plt.subplots_adjust(wspace=0.5, hspace=0.5) # add paddings between subplots
for i in range(6):
    plt.subplot(2, 3, i+1) # 2-rows 3-columns, (i+1)th subplot
    plt.title(i+1) # add title to subplot
    plt.plot(x, f(x, i), color='blueviolet') # plot
    plt.ylim(-20, 20) # set y limit
    plt.grid(True)
plt.show()
```
