```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## Expotential and Logarithm
* Expotential function
$$ f(x)=e^x $$
* Logarithm function
$$ f(x) = \ln{x} $$

```python
x = np.linspace(-5, 5, 100)
plt.plot(x, np.exp(x), color='blueviolet', label='$e^x$')
plt.plot(x, np.log(x), color='magenta', label='$\ln{x}$')
plt.plot(x, x, color='black', label='$y=x$')
plt.ylim(-3, 3)
plt.grid(True)
plt.legend(loc='upper left')
plt.show()
```

## Using logarithm in machine learning

$$\log{\prod_{n=1}^{N} f(n) = \sum_{n=1}^{N} \log{f(n)}}$$

* The key thing is that critical points does not change
$$\min_{x}f(x) = \min_{x} \log{f(x)}$$

ex)
$$f(x) = (x-2)^2 $$

```python
def f(x):
    return np.power(x-2, 2)

plt.plot(x, f(x), color='blueviolet', label='$f(x)$')
plt.plot(x, np.log(f(x)), color='magenta', label='$\log{f(x)}$')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

## Sigmoid
$$\sigma(x)=\frac{1}{1+\exp{(-x)}}$$

* Derivative of Sigmoid
$$\sigma^{'}(x)=\sigma(x)(1-\sigma(x))$$

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def dev_sigmoid(x):
    sig_x = sigmoid(x)
    return np.multiply(sig_x, 1 - sig_x)

x = np.linspace(-10, 10, 100)

plt.plot(x, sigmoid(x), color='blueviolet', label='$\sigma(x)$')
plt.plot(x, dev_sigmoid(x), color='magenta', label="$\sigma^{'}(x) $")
plt.legend(loc ='upper left')
plt.grid(True)
plt.show()
```

## Softmax
$$y_k=\frac{\exp(x_k)}{\sum_{k=1}^{N}{\exp(x_k)}}$$

express number as probability maintaining the reletive size

```python
def softmax(x):
    exp_x = np.exp(x)
    return exp_x / np.sum(exp_x)

n_pts = 5
idx = np.linspace(0, 1, n_pts, endpoint=False)
x = np.random.rand(n_pts, 1)

plt.subplots_adjust(hspace=0.5)

plt.subplot(2, 1, 1)
plt.title('$x$')
plt.plot(idx, x, color='blueviolet')

plt.subplot(2, 1, 2)
plt.title('softmax')
plt.plot(idx, softmax(x), color='magenta')

plt.show()
```

## Sigmoid and softmax

the derivative of softmax is
$$\frac{\partial y_j}{\partial x_i}=y_j(I_{ij}-y_i)$$
which is very similar to $\sigma^{'}$
$$\sigma^{'}(x)=\sigma(x)(1-\sigma(x))$$

## Relationship between sigmoid and softmax

consider two variable version of softmax
$$y = \frac{\exp(x_1)}{\exp(x_1) + \exp(x_2)}$$

multiply $-\exp(x_1)$ to numerator and denominator

$$y = \frac{1}{1 + \exp(-(x_1-x_2))}$$

substitute $x=x_1-x_2$
$$y=\frac{1}{1+\exp(-x)}$$

we can see that softmax is multivariable version of sigmoid
